# Lending Club Case Study

# Understanding the Problem and the Data

### Objective: 

The case study aims to identify risky loan applicants.

### Risky Applicants: 

Risky loan applicants are those who refuse to pay their loans, labeled as "Charged-Off," indicating they are defaulters.

### Aim: 

The study focuses on identifying factors that may contribute to loan applicants defaulting on their payments.

### Import the loan dataset

```python
# import libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', 5000)
pd.set_option('display.max_columns', 5000)
pd.set_option('display.width', 10000)


# load dataset
loan_df = pd.read_csv("loan.csv", low_memory=False);
```

```python
cols_to_keep = ['loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'term', 'int_rate', 
                'installment', 'grade', 'sub_grade' , 'emp_length', 'home_ownership', 
                'annual_inc', 'verification_status', 'loan_status', 'pymnt_plan', 'purpose', 
                'title', 'dti', 'earliest_cr_line', 'inq_last_6mths', 'open_acc', 'pub_rec', 
                'total_acc', 'pub_rec_bankruptcies', 'issue_d']
loan_df = loan_df.loc[:, cols_to_keep]
print(loan_df.shape)
```

### NA values after removing columns with > 30% missing values in clean dataset

```python
missing_vals = (loan_df.isna().mean()*100).sort_values(ascending=False);
print(missing_vals);
```

### Handle missing values for selected columns

```python
# in emp_length column
loan_df['emp_length'] = loan_df['emp_length'].fillna(loan_df['emp_length'].mode()[0])

# pub_rec_bankruptcies
loan_df['pub_rec_bankruptcies'] = loan_df['pub_rec_bankruptcies'].fillna(loan_df['pub_rec_bankruptcies'].median())

# title
loan_df['title'] = loan_df['title'].fillna(loan_df['title'].mode()[0])
```

### Select columns with datatype = object and inspect their sample data

```python
obj_cols = loan_df.select_dtypes(include=["object"]).columns;

for ocol in obj_cols:
    print('---------------------------------------------------------------------------------------------------------\n');
    print('sample data in column ', ocol , '\n');
    print(loan_df[ocol].head(n=10));

```

### Perform Data Transformation
Convert datatypes

```python
import re
# term column
loan_df['term'] = loan_df['term'].apply(lambda x : re.findall(r'\d+', x)[0]).astype('int64');
```

```python
# int_rate Column
loan_df['int_rate'] = loan_df['int_rate'].apply(lambda x : str(x).replace('%', '')).astype('float64');
```

```python
# emp_length column
# Employment length in years. Possible values are between 0 and 10 where 0 means less than one year and 10 means ten or more years. 
# Needs data correction
loan_df['emp_length'] = loan_df['emp_length'].apply(lambda x : str(x).replace('< 1 year', '0').replace('10+ years', '10').replace('years', '').replace('year', '').strip()).astype('int64')
print(loan_df['emp_length'].value_counts());
```

```python
import datetime
loan_df['issue_d_month'] = loan_df['issue_d'].apply(lambda x : datetime.datetime.strptime(str(x).replace('-', ' 20'),'%b %Y').strftime('%m')).astype('int64')
loan_df['issue_d_year'] = loan_df['issue_d'].apply(lambda x : datetime.datetime.strptime(str(x).replace('-', ' 20'),'%b %Y').strftime('%Y')).astype('int64')
```

### Heatmap Analysis - Mean Loan Amount by Loan Status and Purpose

```python
# Heatmap for loan amount, loan status, and loan purpose.
pt = non_current.pivot_table(values='loan_amnt', index='loan_status', columns='purpose', aggfunc='mean')
plt.figure(figsize=(13, 5))
sns.heatmap(pt, cmap='coolwarm',cbar_kws={"label": "Mean Loan Amount"})

plt.title('Mean Loan Amount by Loan Status and Purpose')
plt.xlabel('Loan Purpose', fontsize=14)
plt.ylabel('Loan Status', fontsize=14)
plt.show()
```

### Pairplot Analysis

```python
# Calculate the correlation matrix for non_current
# numeric_columns = ['loan_amnt','int_rate','annual_inc','dti']
numeric_columns = ['loan_amnt', 'annual_inc', 'funded_amnt', 'term', 'int_rate', 'installment']
correlation_matrix = non_current[numeric_columns].corr()
print("Correlation Matrix:\n", correlation_matrix)

# Create a pairplot for the numeric columns in the non_current DataFrame
sns.pairplot(non_current[numeric_columns])
plt.suptitle('Pairplot of Numeric Variables', y=1.02, fontsize=20)
plt.show()
```

# Summary

### Univariate Analysis:
Applicants with rented homes accounted for the highest defaults (50%).
Debt consolidation loans led to the most defaults (49%).
Borrowers with an interest rate between 11% to 17% had the highest default rate.
Loans in the amount range of 5000 to 10000 were most prone to default.

### Derived Metrics:
Type-driven Metric: Applicants with 10+ years of experience account for 28% of defaults, the highest among experience groups.
Business-driven Metric: Loans issued in December have the highest default rate.
Data-driven Metrics:
Applicants with an annual income between 30,000 and 60,000 defaulted the most.
Interestingly, those in the lower income range (1,000 to 30,000) showed fewer defaults.
Borrowers with a monthly debt-to-income ratio of 10% to 20% had the highest default rate.

### Bivariate Analysis:
Annual Income vs Loan Purpose: The highest default rates are seen in loans for credit cards, small businesses, home improvement, and house mortgages.
Annual Income vs Debt-to-Income (DTI) Ratio: Borrowers with incomes between 30000 and 70000 and a DTI ratio between 15% and 25% show debt management issues.
Loan Amount vs Annual Income: A moderate positive correlation (0.41) indicates that as annual income increases, the loan amount tends to rise as well.
Loan Amount vs Grade: High default rates are observed when the loan grade is F or G and the loan amount exceeds 17000.
